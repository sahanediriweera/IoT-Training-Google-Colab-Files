```python
from pytube import YouTube
import os

def downloadYouTube(videourl, path):

    yt = YouTube(videourl)
    yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
    if not os.path.exists(path):
        os.makedirs(path)
    yt.download(path)
```

```python
from os import path
from turtle import width
import torch 
import numpy as np 
import cv2
import matplotlib.pyplot as plt
import os
import scipy.misc
from zmq import device

model = torch.hub.load('ultralytics/yolov5','custom', '/content/weights/best.pt')

cap = cv2.VideoCapture('vid5.mp4')

height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
now_frame = 0
total_fps = fps*116

video_writer = cv2.VideoWriter(os.path.join('data','output.avi'),cv2.VideoWriter_fourcc('P','I','M','1'),fps,(width,height))

while cap.isOpened():
    ret, frame = cap.read()
    if frame.all() == None:
      break
    results = model(frame)

    

    video_writer.write(np.squeeze(results.render()))
    print((now_frame/total_fps)*100)
    now_frame +=1

    

cap.release()
cv2.destroyAllWindows()
video_writer.release()
```
