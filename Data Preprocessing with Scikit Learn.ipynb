```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
df.columns
```

```python
select_df = df[['MSSubClass',	'MSZoning'	,'LotFrontage','LotFrontage','LotArea','Street','LotShape','LandContour','Utilities','MiscVal','MoSold','YrSold',
                'SaleType','SalePrice','SaleCondition']].dropna()
```

```python
X = pd.get_dummies(select_df.drop('SalePrice',axis = 1))
y = select_df.SalePrice
```

```python
pipeline = make_pipeline(StandardScaler(),RandomForestRegressor())
```

```python
pipeline.fit(X,y)
```

```python
import pickle
```

```python
with open('pipelinepickle.pkl','wb') as f:
  pickle.dump(pipeline,f)
```

```python
with open('pipelinepickle.pkl','rb') as f:
  reloaded_model = pickle.load(f)
```
